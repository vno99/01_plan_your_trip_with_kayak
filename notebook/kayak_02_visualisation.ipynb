## 02 - Visualisation

Import de librairies

```python
import pandas as pd
import plotly.express as px
from dotenv import load_dotenv
from IPython.display import Image

load_dotenv()
```

```python
df_hotels = pd.read_csv("../data/scrap_booking.csv")
```

```python
data_hotels = df_hotels.copy()
data_hotels.head()
```

```python
all_villes_score = pd.read_csv("../data/cities.csv")
```

## Cartes

```python
top5_villes = all_villes_score[:5]
print(top5_villes)

top20_hotels = data_hotels[data_hotels["city"].isin(top5_villes["city"])]
print(top20_hotels)
```

```python
def show_graph(fig, filename="map.png"):
    
    fig.write_image(filename)

    try:
        from IPython import get_ipython
        if get_ipython() is not None:
            fig.show()
        else:
            display(Image(filename))
    except:
        display(Image(filename))
```

```python
top5_villes_map = px.scatter_geo(
    top5_villes,
    lat="lat",
    lon="lon",
    hover_name="city",
    size="score_normalized",
    color="score_normalized",
    color_continuous_scale="thermal",
    projection="mercator",
    scope="europe",
    title="Top 5 des villes à visiter",
    size_max=40,
    labels={"score_normalized": "Score"},
    width=1000,
    height=800,
)

top5_villes_map.update_layout(
    geo=dict(
        center=dict(lat=46.5, lon=2.5),  # centre approximatif sur la France
        lataxis=dict(range=[41, 51]),    # limites latitude
        lonaxis=dict(range=[-5, 10])     # limites longitude
    )
)

top5_villes_map.show()

top5_villes_filename = "../images/top5_villes_map.png"
top5_villes_map.write_image(top5_villes_filename)
display(Image(top5_villes_filename))
```

```python
top20_hotels_map = px.scatter_geo(
    top20_hotels,
    lat="lat",
    lon="lon",
    hover_name="name",
    hover_data=["city", "lat", "lon", "review_score"],
    size="review_score",
    color="review_score",
    color_continuous_scale="thermal",
    projection="mercator",
    scope="europe",
    title="Top 20 des hotels par ville",
    labels={"city": "Ville"},
    width=1000,
    height=800,
)

top20_hotels_map.update_layout(
    geo=dict(
        center=dict(lat=46.5, lon=2.5),  # centre approximatif sur la France
        lataxis=dict(range=[41, 51]),    # limites latitude
        lonaxis=dict(range=[-5, 10])     # limites longitude
    )
)

top20_hotels_map.show()

top20_hotels_filename = "../images/top20_hotels_map.png"
top20_hotels_map.write_image(top20_hotels_filename)
display(Image(top20_hotels_filename))
```
